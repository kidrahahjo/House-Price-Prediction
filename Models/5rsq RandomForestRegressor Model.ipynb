```python
import pandas as pd
import pandasql as pdsql
import seaborn as sns
import signac

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_log_error

sns.set_theme(style="whitegrid")
```

```python
project = signac.get_project('dataset_per_rsquared')
```

```python
project
```

```python
for job in project:
    if job.sp.get('5%') is not None:
        req_job = job
        break
```

```python
req_job
```

```python
features =req_job.sp.get('5%')
features
```

```python
dataset_train = pd.read_csv("../5rsq_final_data.csv")
dataset_test = pd.read_csv("../Data/test.csv")[['Id'] + features[:-1]]
```

```python
categorical_cols = []
for column in features[:-1]:
    if str(dataset_train[column].dtype) == 'object':
        categorical_cols.append(column)
        dataset_test[column] = dataset_test[column].fillna('None')
        dataset_train[column] = dataset_train[column].fillna('None')
        labelencoder = LabelEncoder()
        dataset_test[column] = pd.Series(labelencoder.fit_transform(dataset_test[column]))
        dataset_train[column] = pd.Series(labelencoder.fit_transform(dataset_train[column]))

non_cat_cols = list(set(features).difference(set(categorical_cols)))
```

```python
dataset_test = dataset_test.fillna(-1)
dataset_train = dataset_train.fillna(-1)
```

```python
X = dataset_train[features[:-1]]
y = dataset_train['SalePrice']
test_X = dataset_test[features[:-1]]
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2)
```

```python
X.head()
```

```python
y.head()
```

```python
dataset_test.head()
```

```python
regr = RandomForestRegressor()
regr.fit(train_X, train_y)
val_r2_score = regr.score(val_X, val_y)
val_pred = regr.predict(val_X)
val_msle_score = mean_squared_log_error(val_y, val_pred)
```

```python
val_r2_score
```

```python
val_msle_score
```

```python
final_predict = regr.predict(test_X)
```

```python
final_predict = pd.DataFrame({'Id': list(dataset_test['Id']), 'SalePrice': final_predict, })
final_predict
```

```python
final_predict.to_csv('../Submissions/5r2sample_rfr.csv', index=False)
```

```python
req_job.sp['val_mean_squared_log_error'] = val_r2_score
req_job.sp['submit_mean_squared_log_error'] = val_msle_score
```
