## R-squared cleaning
We will first analyse how much each column explains the variance in `SalePrice` and then reject variables according to a specific minimum value

```python
rsquared_dataset = clean_data.analyse_rsquared(dataset)
rsquared_dataset.head()
```

```python
def show_values_on_bars(graph, total):
    "Show actual values on plot"
    for p in ax.patches:
        height = p.get_height()
        graph.text(p.get_x() + p.get_width() / 2.,
                height + 3,
                '{:1.2f}'.format((height / 1.23) / total),
                ha="center")
    show()
    
_, ax = plt.subplots(figsize=(100, 16))
total = len(rsquared_dataset['label'])
graph = sns.barplot(x='label', y='variance_explained', data=rsquared_dataset, ax=ax)
show_values_on_bars(graph, total)
```
